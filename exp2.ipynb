```python
import matplotlib.pyplot as plt

sns.set(style='white', palette="Set2", color_codes=False)
sns.set_style("ticks")

colors = sns.color_palette()  #  Set26

conditions = ['Solo Coding', 'Classical Music', 'Workout Before', 'Coding w/ Peer', 'Late Night', 'YouTube', 'Pomodoro', 'ChatGPT', '3 hours before ddl']
efficiency = [100, 145, 110, 85, 60, 20, 150, 210, 300]

# 

plt.figure(figsize=(12, 6))
bars = plt.bar(conditions, efficiency, color='skyblue', edgecolor='black')

# 
for bar in bars:
    yval = bar.get_height()
    plt.text(bar.get_x() + bar.get_width()/2, yval + 2, f'{yval}%', ha='center', va='bottom', fontsize=11)

plt.axhline(100, color='red', linestyle='--', label='Baseline')
plt.ylabel('Relative Efficiency (%)')
plt.title('Effect of Different Conditions on My Productivity', fontsize=16, fontweight='bold')
plt.xticks(rotation=35, ha='right', fontsize=14)
plt.tight_layout()
plt.legend()
plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

#  Seaborn 
sns.set(style='whitegrid', palette='Set2')
plt.rcParams.update({'font.size': 13})

# 
conditions = ['Solo Coding', 'Classical Music', 'Workout Before', 'Coding w/ Peer', 'Late Night', 'YouTube', 'Pomodoro', 'ChatGPT', '3 hours before ddl']
efficiency = [100, 145, 110, 85, 60, 20, 150, 210, 300]

#  DataFrame
df_eff = pd.DataFrame({
    'Condition': conditions,
    'Relative Efficiency (%)': efficiency
})

# 
max_eff = max(efficiency)
max_index = efficiency.index(max_eff)
max_condition = conditions[max_index]

# 
plt.figure(figsize=(13, 6))
barplot = sns.barplot(
    x='Condition',
    y='Relative Efficiency (%)',
    data=df_eff,
    palette='Set2',
    edgecolor='black'
)

# 
for i, p in enumerate(barplot.patches):
    height = p.get_height()
    barplot.annotate(
        f'{height}%',
        (p.get_x() + p.get_width() / 2., height),
        ha='center', va='bottom',
        fontsize=11
    )
    # 
    if i == max_index:
        barplot.annotate(
            'SOTA',
            (p.get_x() + p.get_width() / 2., height + 12),
            ha='center', va='bottom',
            fontsize=13, fontweight='bold', color='darkred'
        )


# 
plt.axhline(100, color='red', linestyle='--', linewidth=1.2, label='Baseline (Solo Coding)')

# 
plt.title("Effect of Conditions on Haibin's Productivity", fontsize=18, pad=12, fontweight='bold')
plt.ylabel("Relative Efficiency (%)", fontsize=14, fontweight = 'bold')
plt.xticks(rotation=25, ha='right', fontsize=14, fontweight = 'bold') # bold
plt.legend()

# 
plt.tight_layout()

# 
plt.savefig('productivity_conditions_seaborn.png', dpi=300, bbox_inches='tight')

# 
plt.show()
```
