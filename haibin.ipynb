```python
import matplotlib.pyplot as plt
import seaborn as sns

hours = {
    "Reading Papers": 8,
    "Coding": 18, 
    "Benchmarking": 10,
    "Debugging (!)": 4
}

# 设置HPC主题色（蓝紫科技感）
colors = ["#4C72B0", "#DD8452", "#55A868", "#C44E52"]
explode = (0, 0.1, 0, 0)  # 突出Coding部分

plt.figure(figsize=(8, 6))
plt.pie(hours.values(), labels=hours.keys(), autopct='%1.1f%%',
        startangle=90, colors=colors, explode=explode,
        wedgeprops={'linewidth': 1, 'edgecolor': 'white'})

# 添加标题和学术标注
plt.title("Weekly Research Hours Distribution\n(Total: 40±5 hours)", 
          pad=20, fontdict={'fontsize':14, 'weight':'bold'})
plt.text(-1.5, -1.2, "Data collected via Toggl Track", 
         fontsize=9, alpha=0.7)

plt.tight_layout()
plt.savefig('research_hours_pie.png', dpi=300, bbox_inches='tight')
```

```python
# 转换为DataFrame
import pandas as pd
df = pd.DataFrame.from_dict(hours, orient='index', columns=['Hours'])

# 创建水平条形图
plt.figure(figsize=(10, 4))
ax = sns.barplot(x=df['Hours'], y=df.index, palette=colors, 
                 edgecolor=".2", linewidth=1.5)

# 添加数值标签
for i, (index, row) in enumerate(df.iterrows()):
    ax.text(row.Hours+0.3, i, f"{row.Hours}h", ha='left', va='center')

# 学术风格调整
ax.set_xlabel("Hours per Week", fontweight='bold')
ax.set_ylabel("Activity", fontweight='bold')
sns.despine(left=True, bottom=True)
plt.grid(axis='x', linestyle='--', alpha=0.4)

plt.savefig('research_hours_bar.png', dpi=300, transparent=True)
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

# 示例数据 - 各阶段在不同活动上的时间分配（小时/周）
data = {
    'Phase': ['Planning', 'Finding Failures', 'Design', 'Validation', 'Pruning', 'Writing'],
    'Reading Papers': [4, 6, 3, 2, 1, 2],
    'Benchmarking': [3, 8, 2, 6, 4, 1],
    'Profiling': [2, 5, 4, 8, 6, 1],
    'Coding': [5, 4, 8, 10, 8, 3],
    'Writing Papers': [1, 2, 1, 1, 2, 8]
}

# 转换为长格式
df = pd.DataFrame(data).melt(id_vars='Phase', var_name='Activity', value_name='Hours')

# 设置颜色和样式
sns.set(style="whitegrid", palette="muted")
plt.figure(figsize=(14, 8))

# 创建6个子图，每个阶段一个
phases = data['Phase']
fig, axes = plt.subplots(2, 3, figsize=(18, 10))
axes = axes.flatten()

for i, phase in enumerate(phases):
    phase_data = df[df['Phase'] == phase]
    sns.barplot(x='Activity', y='Hours', data=phase_data, ax=axes[i], palette="Blues")
    axes[i].set_title(f'{phase} Phase', fontsize=12)
    axes[i].set_ylim(0, 12)
    
    # 添加数值标签
    for p in axes[i].patches:
        axes[i].annotate(f"{p.get_height():.0f}", 
                        (p.get_x() + p.get_width() / 2., p.get_height()), 
                        ha='center', va='center', 
                        xytext=(0, 5), 
                        textcoords='offset points')

plt.suptitle('Weekly Time Allocation Across Research Phases', fontsize=16, y=1.02)
plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
import numpy as np

# 设置配色和样式
colors = sns.color_palette()  # 默认使用 Set2，6种颜色
sns.set(style='white', palette="Set2", color_codes=False)
sns.set_style("ticks")

# 原始数据
data = {
    'Activity': ['Reading Papers', 'Benchmarking', 'Profiling', 'Coding', 'Writing Papers'],
    'Planning': [16, 8, 8, 2, 0],
    'Finding Failures': [8, 10, 20, 2, 0],
    'Design': [14, 3, 6, 16, 1],
    'Validation': [2, 6, 8, 22, 1],
    'Pruning': [1, 2, 2, 24, 4],
    'Writing': [7, 4, 3, 3, 20]
}

# 转置数据：x轴为Phase，柱子为Activity
df = pd.DataFrame(data)
df_melted = df.melt(id_vars='Activity', var_name='Phase', value_name='Hours')

# 获取唯一值顺序
phases = df_melted['Phase'].unique()
activities = df_melted['Activity'].unique()
n_activities = len(activities)
n_phases = len(phases)

# 分组柱状图参数
bar_width = 0.17
index = np.arange(len(phases))

# 绘图
fig, ax = plt.subplots(figsize=(15, 5))

# 每种 Activity 画一组柱子
for i, activity in enumerate(activities):
    subset = df_melted[df_melted['Activity'] == activity]
    values = subset['Hours'].values
    ax.bar(index + i * bar_width, values, bar_width, label=activity, color=colors[i], edgecolor='black')

# 设置 x 轴
ax.set_xticks(index + bar_width * (n_activities - 1) / 2)
ax.set_xticklabels(phases, fontsize=17)

# 添加数值标签
for i, activity in enumerate(activities):
    subset = df_melted[df_melted['Activity'] == activity]
    values = subset['Hours'].values
    for j, val in enumerate(values):
        if val > 0:
            ax.text(index[j] + i * bar_width, val + 0.2, f'{val}', ha='center', va='bottom', fontsize=15)

# log y
# ax.set_yscale('log')

# 标题和图例
plt.title('Time Allocation by Research Phase and Activity', fontsize=20, pad=12, fontweight='bold')
# plt.xlabel('Research Phase', fontsize=18)
plt.ylabel('Hours per Week', fontsize=18)
plt.legend(
    # title='Research Activity',
    loc='upper center',
    bbox_to_anchor=(0.5, -0.15),  # 横向居中，纵向下移
    ncol=5,                       # 可根据活动数量调整列数
    frameon=False,
    fontsize=15,
    title_fontsize=12
)


plt.tight_layout()
sns.despine()
# save as png

plt.savefig('research_hours_allocation.png', dpi=300, bbox_inches='tight')

plt.show()
```
