```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

# 
sns.set(style='whitegrid', palette='Set2')
plt.rcParams.update({'font.size': 13})

# 
data = {
    'Activity': ['Reading Papers', 'Benchmarking', 'Profiling', 'Coding', 'Writing Papers'],
    'Planning': [11, 8, 6, 1, 0],
    'Finding Failures': [2, 6, 17, 1, 0],
    'Design': [7, 3, 6, 16, 0],
    'Validation': [2, 6, 8, 22, 1],
    'Pruning': [1, 2, 2, 24, 4],
    'Writing': [7, 4, 3, 3, 20]
}

#  DataFrame 
df = pd.DataFrame(data)
df_stacked = df.set_index('Activity').T

# 
fig, ax = plt.subplots(figsize=(12, 7))
df_stacked.plot(kind='bar', stacked=True, ax=ax, edgecolor='black', width=0.7)

# 
ax.set_title("Haibin's Weekly Time Allocation by Research Phase", fontsize=18, pad=15, fontweight='bold')
ax.set_xlabel('Research Phase', fontsize=14)
ax.set_ylabel('Weekly Hours', fontsize=14)
ax.legend(title='Activity', bbox_to_anchor=(1.00, 1), loc='upper left', frameon=False, fontsize=12)

# # 
# for i, phase in enumerate(df_stacked.index):
#     total = df_stacked.loc[phase].sum()
#     ax.text(i, total + 1, f'{int(total)}h', ha='center', va='bottom', fontsize=12)

# 
for bar_idx, phase in enumerate(df_stacked.index):
    cumulative = 0
    max_val = 0 
    for act_idx, val in enumerate(df_stacked.loc[phase]):
        if val > max_val:
            max_val = val
        # cumulative += val
        # 
        
    for act_idx, val in enumerate(df_stacked.loc[phase]):
        if val > 0 and val == max_val:
            ax.text(bar_idx, cumulative + val / 2, f'{int(val)}h', ha='center', va='center', fontsize=12, fontweight='bold')
        cumulative += val
    # 
    ax.text(bar_idx, cumulative + 1, f'{int(cumulative)}h', ha='center', va='bottom', fontsize=13, fontweight='bold')

# 
ax.grid(False)

# x 
ax.set_xticklabels(df_stacked.index, rotation=0, ha='center', fontsize=14)

sns.despine()
plt.tight_layout()

# 
plt.savefig('haibin_stacked_research_hours.png', dpi=300, bbox_inches='tight')
plt.show()
```
